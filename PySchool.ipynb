```python
import pandas as pd
```

```python
#Import the files and create a DataFrame
file1="Resources/schools_complete.csv"
file2="Resources/students_complete.csv"
#School dataframe
school_df=pd.read_csv(file1)
school_df
```

```python
#Students dataframe
student_df=pd.read_csv(file2)
student_df
# student_df.count()
```

```python
#merging two data files and create one complete school data
school_data_complete= pd.merge(school_df,student_df,on="school_name", how="left")
school_data_complete
```

```python
#Total no of schools
school_total= school_df["School ID"].count()
school_total
```

```python
#Total no of students
student_total= student_df["Student ID"].count()
student_total
```

```python
#Total Budget
budget_total= school_df["budget"].sum()
budget_total
```

```python
#Average Math Score
average_math_score= student_df["math_score"].mean()
round(average_math_score,3)
```

```python
#Average Reading score
average_reading_score=student_df["reading_score"].mean()
round(average_reading_score,3)
```

```python
#% Passing math score
math_score_passed= school_data_complete.loc[school_data_complete["math_score"]>=70].count()
perct_passed_math=(math_score_passed["math_score"]/ student_total)*100
round(perct_passed_math,2)
```

```python
# % Passing reading score
reading_score_passed= school_data_complete.loc[school_data_complete["reading_score"]>=70].count()
perct_passed_reading=(reading_score_passed["reading_score"]/student_total)*100
round(perct_passed_reading,3)
```

```python
#Calculate overall passing percentage
both_pass = student_df.loc[((student_df.reading_score) >=70) & ((student_df.math_score) >=70)]
overall_pass=((both_pass['student_name'].count())/student_total)*100
round(overall_pass,3)
```

```python
district_summary= pd.DataFrame({'Total Schools':[school_total],
                                'Total Students':[student_total],
                                'Total Budget':[budget_total],
                                'Average Math Score':[average_math_score],
                                'Average Reading Score':[average_reading_score],
                                '%Passing Math':[perct_passed_math],
                                '%Passing Reading':[perct_passed_reading],
                                '%Overall Passing':[overall_pass]})

district_summary
```

```python
district_student_summary_df=district_summary[['Total Schools',
                                'Total Students',
                                'Total Budget',
                                'Average Math Score',
                                'Average Reading Score',
                                '%Passing Math',
                                '%Passing Reading',
                                '%Overall Passing']]
district_student_summary_df=district_student_summary_df.round(2)
```

```python
#Formatting the District Summary df
district_student_summary_df["%Passing Math"] = district_student_summary_df["%Passing Math"].map("{:.2f}%".format)
district_student_summary_df["Total Students"]=district_student_summary_df["Total Students"].map("{:,.0f}".format)
district_student_summary_df["Total Budget"]=district_student_summary_df["Total Budget"].map("${:,.2f}".format)
district_student_summary_df["%Passing Reading"]=district_student_summary_df["%Passing Reading"].map("{:.2f}%".format)
district_student_summary_df["%Overall Passing"]=district_student_summary_df["%Overall Passing"].map("{:.2f}%".format)

district_student_summary_df
```

```python
#Create School Summary
# Set school name as index 
group_by_district= school_df.set_index(['school_name'])['type']
group_by_district
```

```python
#Total students in each school
student_count= school_data_complete['school_name'].value_counts()
student_count
```

```python
#total budget for each school
school_budget= school_data_complete.groupby(['school_name'])['budget'].mean()
school_budget
```

```python
# Budget based on per student in each school
per_student_budget= school_budget/student_count
per_student_budget
```

```python
#Average Math Score in each school
avg_math_score= school_data_complete.groupby(['school_name'])['math_score'].mean()
avg_math_score
```

```python
#Average Reading score in each school
avg_reading_score= school_data_complete.groupby(['school_name'])['reading_score'].mean()
avg_reading_score
```

```python
#Passing Math % in each school
math_score_byschools= school_data_complete.loc[school_data_complete['math_score']>=70]
group_by_school=math_score_byschools['school_name'].value_counts()
passing_math_perct= (group_by_school/student_count)*100
passing_math_perct
```

```python
#Passing Reading % in each school
reading_score_byschools= school_data_complete.loc[school_data_complete['reading_score']>=70]
group_by_school=reading_score_byschools['school_name'].value_counts()
passing_reading_perct= (group_by_school/student_count)*100
passing_reading_perct
```

```python
#Overall passing % in each school
math_reading_pass = school_data_complete.loc[((school_data_complete.reading_score) >=70) & ((school_data_complete.math_score) >=70)]
group_math_reading=math_reading_pass['school_name'].value_counts()
overall_pass_perct=(group_math_reading/student_count)*100
overall_pass_perct
    
    
```

```python
#Display the School Summary report
school_summary=pd.DataFrame({"School Type":group_by_district,
                              "Total Students":student_count,
                              "Total School Budget":school_budget,
                              "Per Student Budget":per_student_budget,
                              "Average Math Score":avg_math_score,
                              "Average Reading Score":avg_reading_score,
                              "% Passing Math":passing_math_perct,
                              "% Passing Reading":passing_reading_perct,
                              "% Overall Passing":overall_pass_perct})
school_summary
```

```python
school_summary_df=school_summary[["School Type",
                              "Total Students",
                              "Total School Budget",
                              "Per Student Budget",
                              "Average Math Score",
                              "Average Reading Score",
                              "% Passing Math",
                              "% Passing Reading",
                              "% Overall Passing"]]
school_summary_df=school_summary_df.round(2)
```

```python
#Formatted School Summary
school_summary_df["Total School Budget"]= school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"]= school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["% Passing Math"]= school_summary_df["% Passing Math"].map("{:.2f}%".format)
school_summary_df["% Passing Reading"]= school_summary_df["% Passing Reading"].map("{:.2f}%".format)                                                                                  
school_summary_df["% Overall Passing"]= school_summary_df["% Overall Passing"].map("{:.2f}%".format)
school_summary_df
```

```python
#Top 5 performing schools by overall passing
top_five_schools= school_summary_df.sort_values("% Overall Passing", ascending=False)
top_five_schools.head(5)
```

```python
#Bottom 5 performing schools
bottom_five_schools= school_summary_df.sort_values("% Overall Passing", ascending=True)
bottom_five_schools.head()
```

```python
#Math Scores by Grade
grade_9th_math= school_data_complete.loc[school_data_complete['grade']=='9th'].groupby(['school_name'])['math_score'].mean()
grade_10th_math= school_data_complete.loc[school_data_complete['grade']=='10th'].groupby(['school_name'])['math_score'].mean()
grade_11th_math= school_data_complete.loc[school_data_complete['grade']=='11th'].groupby(['school_name'])['math_score'].mean()
grade_12th_math= school_data_complete.loc[school_data_complete['grade']=='12th'].groupby(['school_name'])['math_score'].mean()


math_score_bygrade=pd.DataFrame({"9th":grade_9th_math,
                                  "10th":grade_10th_math,
                                  "11th":grade_11th_math,
                                  "12th":grade_12th_math})
math_score_bygrade.round(2)




```

```python
#Reading score by Grade
grade_9th_read= school_data_complete.loc[school_data_complete['grade']=='9th'].groupby(['school_name'])['reading_score'].mean()
grade_10th_read= school_data_complete.loc[school_data_complete['grade']=='10th'].groupby(['school_name'])['reading_score'].mean()
grade_11th_read= school_data_complete.loc[school_data_complete['grade']=='11th'].groupby(['school_name'])['reading_score'].mean()
grade_12th_read= school_data_complete.loc[school_data_complete['grade']=='12th'].groupby(['school_name'])['reading_score'].mean()
reading_score_bygrade=pd.DataFrame({'9th':grade_9th_read,
                                    '10th':grade_10th_read,
                                    '11th':grade_11th_read,
                                    '12th':grade_12th_read})
reading_score_bygrade.round(2)
```

```python
#Scores by school spending
#creating bins for the range of spending
bins=[0, 585, 629, 644, 675]
labels=['<$584','$585-629','$630-644','$645-675']
```

```python
school_summary['Spending Range(Per Student)']=pd.cut(school_summary['Per Student Budget'], bins, labels=labels)
```

```python
#Score by School spending per student display
new_spending_df=school_summary.drop(columns=['School Type','Total Students','Total School Budget','Per Student Budget'])
round(new_spending_df.groupby(["Spending Range(Per Student)"]).mean(),2)
```

```python
#Score by School Size
bins1=[0, 1000, 2000, 5000]
labels_size=['Small(<1000)','Medium(1000-2000)','Large(2000-5000)']
```

```python
#Display score by school size
school_summary['School Size']=pd.cut(school_summary['Total Students'], bins1, labels=labels_size)
new_score_bysize_df= school_summary.drop(columns=['School Type','Total Students','Total School Budget','Per Student Budget','Spending Range(Per Student)'])
round(new_score_bysize_df.groupby(['School Size']).mean(),2)



```
